# Convert text into chunks and explain the logic of chunking strategy

## Chunking Strategy: OCR Extracted Text is saved in structured JSON format.
- Every text is grouped across chunk_size = 300 and an overlap of 50 tokens - Word split (tokens) is considered.
- The data has unnecessary spaces, literals, required a cleaning structure, appended the chunks with overlap chunks post cleaned.
- we are persisting the json data that is chunked.

```python
import json

# Load the extracted text data from the JSON file
with open(txtpath+'/ocr_extracted_text.json', 'r') as f:
    cleaned_text_data = json.load(f)


```

```python
# Define the chunking logic
def chunk_text(text, chunk_size=300, overlap=50):
    words = text.split()
    chunks = []
    start = 0
    while start < len(words):
        end = start + chunk_size
        chunk = words[start:end]
        chunks.append(' '.join(chunk))
        start += chunk_size - overlap
    return chunks
```

```python

# Apply the chunking logic to the extracted text
all_chunks = []
for text in cleaned_text_data:
    chunks = chunk_text(text, chunk_size=500, overlap=50)
    all_chunks.extend(chunks)
```

```python
# Store the text chunks in JSON format
with open(txtpath+'/text_chunks.json', 'w') as f:
    json.dump(all_chunks, f)
print("Text chunking completed successfully.")
```

1. Data Ingestion
Objective:

Objective:
•	To extract raw text data from the loaded PDF files.

•	Tools and Libraries: Leverage any open source OCR extraction library to extract the content from unstructured document. We are using tesseract library
3. Data Preprocessing
Objective:
•	To clean and preprocess the extracted text data for further analysis.
Details:
•	The raw text extracted from PDF files often contains noise such as special characters, extra spaces, and stop words that are not useful for analysis. Preprocessing involves steps like tokenization, converting text to lowercase, removing punctuation and stop words, and stemming or lemmatization. This step ensures that the data is in a structured format suitable for analysis or further processing.
4. Data – Chunks Vector Embeddings

## Enhancement : Customizing the stages of querying through LLAMAINDEX Response Synthesizer

- We customize our retriever to use a different number for top_k and add a post-processing step that requires that the retrieved nodes reach a minimum similarity score to be included. Here it is. 0.7
-  This would give you more richer, relevant results but potentially no data if you have nothing relevant.

```python
from llama_index.core import VectorStoreIndex, get_response_synthesizer
from llama_index.core.retrievers import VectorIndexRetriever
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.core.postprocessor import SimilarityPostprocessor


# configure retriever
retriever = VectorIndexRetriever(
    index=index,
    similarity_top_k=10,
)

# configure response synthesizer
response_synthesizer = get_response_synthesizer(llm=llm_model)

# assemble query engine
query_engine = RetrieverQueryEngine(
    retriever=retriever,
    response_synthesizer=response_synthesizer,
    node_postprocessors=[SimilarityPostprocessor(similarity_cutoff=0.7)],
)

```

```python
# query
response = query_engine.query("What is HIV Test Kits flow process")
print(response)
```

```python
# query
response_5 = query_engine.query("Brief the scope of Receive and Store Commodities in less than 100 words")
print(response_5)
```
